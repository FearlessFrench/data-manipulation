So far we've imported the modules we're going to use, and defined our two images, the template (img1) and the image we're going to search for the template in (img2)

```python
#cv2.imread(filename, flag)  https://www.geeksforgeeks.org/python-opencv-cv2-imread-method/
#flag: The flag specifies the way how the image should be read.

#cv2.IMREAD_COLOR – It specifies to load a color image. Any transparency of image will be neglected. It is the default flag.
#Alternatively, we can pass integer value 1 for this flag.

#cv2.IMREAD_GRAYSCALE – It specifies to load an image in grayscale mode. Alternatively, we can pass integer value 0 for this flag.

#cv2.IMREAD_UNCHANGED – It specifies to load an image as such including alpha channel. Alternatively, we can pass integer value -1 for this flag.
```

```python
img1 = cv2.imread('opencv-feature-matching-template.jpg',0) # 0 มีความหมายว่า Grayscale
img2 = cv2.imread('opencv-feature-matching-image.jpg',0)
```

```python
cv2_imshow(img1)
cv2_imshow(img2)
```

# ORB (Oriented FAST and Rotated BRIEF)

BRIEF (Binary Robust Independent Elementary Features)

is a feature descriptor, it speeds up computation of descriptors.

```python
#BRIEF (ORB) is a fast robust local feature detector,เป็นเครื่องตรวจจับคุณลักษณะภายในที่รวดเร็ว
```

```python
orb = cv2.ORB_create()
```

This is the detector we're going to use for the features.

```python
# find the keypoints and descriptors
kp1, des1 = orb.detectAndCompute(img1,None) #  queryImage
kp2, des2 = orb.detectAndCompute(img2,None) #  trainImage
```

# Brute Force Matching(BFMatcher)

This is our BFMatcher object.

**Norm Types:**

> cv.NORM_L2 : For SIFT, SURF

> cv.NORM_HAMMING : ORB, BRIEF, BRISK, etc.

> cv.NORM_HAMMING2 : ORB if WTA_K is 3 or 4

```python
# Brute-Force Matching เป็น Matching โดยอาศัยการ Match โดยอาศัยระยะที่น้อยที่สุดใน key point แต่ละชุด จากนั้นเลือกแสดงเฉพาะ Key Point ที่ใกล้เคียงกันเท่านั้น
# https://docs.opencv.org/4.x/dc/dc3/tutorial_py_matcher.html

#NormType specifies the distance measurement to be used.
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
```

```python
matches = bf.match(des1,des2)
matches = sorted(matches, key = lambda x:x.distance)
```

Here we create matches of the descriptors, then we sort them based on their distances.

```python
img3 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:10],None, flags=2)
plt.imshow(img3)
plt.show()
```
