```python
import pandas as pd
import csv
f = open('stu_grade.txt')
reader = csv.reader(f)
for line in reader:
  print(line)
```

```python
data={'Name':['Christopher','Jenny','Jakarine','Alan','Ben','Peter','Mike','Silvia'],
      'Program':['C','Python','Java','Python','Java','C','Python','Java'],
      'G.P.A.':[2.41,3.01,2.73,3.21,2.37,3.23,2.75,2.94],}
```

```python
stu_grade = pd.DataFrame(data)
stu_grade
```

1)  จำนวนของประเภท Program ว่าแต่ละแบบมีนักเรียนทั้งหมดกี่คน และชื่ออะไรบ้ำง

```python
# จำนวนของนักเรียนต่อ Program แต่ละแบบ
count = stu_grade.groupby(['Program']).size()
count
```

```python
# ชื่อของนักเรียนต่อ Program แต่ละแบบ
student = stu_grade.groupby(['Program','Name']).size()
student
```

```python
df = pd.DataFrame({'Program':['C','','','Python','','','','Java','','',''],
                   'Count & Name':[count[0],'Christopher','Peter', count[2],'Jenny','Alan','Mike', count[1],'Jakarine','Ben','Silvia'],})
df
```

2)  รายชื่อของนักเรียนที่มี G.P.A. อยู่ในช่วง 2 - 2.99 และ 3.00 ขึ้นไป

```python
# G.P.A. ช่วง 2.00-2.99
stu_grade[(stu_grade['G.P.A.'] >= 2.00) & (stu_grade['G.P.A.'] <= 2.99)].groupby(['G.P.A.','Name']).size()
```

```python
# G.P.A. ตั้งแต่ 3.00 ขึ้นไป
stu_grade[stu_grade['G.P.A.'] >= 3].groupby(['G.P.A.','Name']).size()
```

```python
res = pd.DataFrame({'G.P.A.':['2.00-2.99','','','','','>=3.00','',''],
                   'Name':['Christopher','Jakarine','Ben','Mike','Silvia','Jenny','Alan','Peter'],})
res
```
