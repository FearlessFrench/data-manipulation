โดยดึงอัตราแลกเปลี่ยนของเงินไทยกับสกุลเงิน USD, JPY และ EUR ในปี 2021 และ ปี 2022 ซึ่งการดึงข้อมูลในแต่ละครั้ง

###a) แสดงอัตราแลกเปลี่ยนของรายวันของเดือนมกราคม มีนาคม พฤษภาคม กรกฎาคม กันยายน พฤศจิกายนของปี 2022 ในรูปแบบของตาราง

```python
#อัตราแลกเปลี่ยนของรายวันของเงินไทยกับสกุลเงิน USD เดือน มกราคม ปี 2022
import http.client
conn = http.client.HTTPSConnection("apigw1.bot.or.th")

headers = { 'X-IBM-Client-Id': "b70cba08-96de-4684-9055-a78d57f73e67" }
conn.request("GET", "/bot/public/Stat-ExchangeRate/v2/DAILY_AVG_EXG_RATE/?start_period=2022-01-01&end_period=2022-01-31&currency=USD", headers=headers)

res = conn.getresponse()
data = res.read()
print(data.decode("utf-8"))
```

```python
# อ่านข้อมูล json ที่อ่านเข้ามาและเก็บไว้ในตัวแปร obj
obj = json.loads(data)
# สร้างตาราง
JAN = pd.DataFrame(obj['result']['data']['data_detail'])
# เลือก columns
JAN_detail = pd.DataFrame(JAN, columns=['period','currency_id','buying_sight','buying_transfer','selling','mid_rate'])
JAN_detail
```
