Regular expressions or regexes

เป็นรูปแบบที่ใช้สำหรับการค้นหา text ที่อยู่ใน string

โดยสามารถใช้ใน standard library's re module ได้

```
# parameter ประกอบด้วย pattern และ string
           # pattern ใช้ wild card ได้
#  * ใช้เพื่อกำหนดการเกิดซ้ำให้กับตัวอักษรด้านหน้าได้ ซึ่ง * เป็นตัวอักษรพิเศษที่หมายถึงให้ดูตัวอักษรด้านหน้า และตั้งเป็น pattern ที่มีการเกิดตัวอักษรซ้ำ หรือไม่เกิดขึ้น
           # findall	Returns a list containing all matches
```

```
import re
# b* หมายถึง b ตั้งแต่ 0 ตัวขึ้นไป (การเกิดตัวอักษรซ้ำ หรือไม่เกิดขึ้นเลย)
# ab* คือ ac  abc  abbc  abbbbc
re.findall("ab*c", "ac")
```

```
string = "Everything is <replaced> if it's in <tags>"
string = re.sub("<.*>","ELEPHANTS",string)
# 'Everything is <replaced> if it's in <tags>'
# เป็น
# 'Everything is ELEPHANTS'
string
```

```
string = "Everything is <replaced> if it's in <tags>"
string = re.sub("<.*?>","ELEPHANTS",string)
# ? คือ เป็นชิ้นๆ

# 'Everything is <replaced> if it's in <tags>'
# เป็น
# 'Everything is ELEPHANTS if it's in ELEPHANTS
string
```

```
import re
from urllib.request import urlopen

url = "http://olympus.realpython.org/profiles/dionysus"
page = urlopen(url)
html = page.read().decode("utf-8")
```

```
print(html)
```

```
pattern = "<title.*?>.*?</title.*?>"
match_results = re.search(pattern, html, re.IGNORECASE)
match_results
```

```
title = match_results.group()
title
```

```
title = re.sub("<.*?>","", title)
# "" คือ แทนที่ด้วยช่องว่าง (การตัด <TITLE > และ </title  / >)
print(title)
```

```
print(html)
```

```
for string in ["Name: ", "Favorite Color:"]:
  string_start_idx = html.find(string)
  text_start_idx = string_start_idx + len(string) # ตรงที่ Name:

  next_html_tag_offset = html[text_start_idx:].find("<")
  text_end_idx = text_start_idx + next_html_tag_offset

  raw_text = html[text_start_idx : text_end_idx]
  clean_text = raw_text.strip(" \r\n\t")
  print(clean_text)
```

# Read html tag from 'http://python.org'

```
import urllib.request

def readWebPage(urlStr):
  assert(urlStr.startswith("http://"))
  filehandler = contents = None
  try:
    filehandler = urllib.request.urlopen(urlStr)
    contents = filehandler.read()
    contents = contents.decode('utf-8')
  finally:
    if (filehandler is not None):
      filehandler.close()
    return contents
```

```
print(readWebPage("http://www.python.org")[:50000])
```
