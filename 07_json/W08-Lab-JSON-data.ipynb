```python
import io
import json
ab={'a':[1,2], "b":(3,4)}
print(ab)
print(json.dumps(ab))  #สังเกตว่าเป็น dumps  ที่เติมตัว s ด้วย จะเป็นการแปลง object เป็นข้อความ json
```

```python
with open('ab.json', 'w') as f:
	json.dump(ab,f) #dump ไม่มีตัว s ต่อท้าย เป็นการเขียน object เป็นข้อความ json ลงในไฟล์ fp (แปลง obj เป็น json ก่อน แล้วค่อยเขียนลงไฟล์)
```

```python
import pandas as pd
data=pd.read_json('ab.json')
data
```

```python
print(json.dumps({(1,2):1})) # ERROR เพราะ ค่า Key เป็น Tuple
```

```python
print(json.dumps({(1,2):1, 3:4},skipkeys=True))
#จะแปลงได้แค่เทอมหลัง คือ 3:4
#ส่วนเทอมแรกที่เป็น  (1,2):1  จะแปลงไม่ได้ เพราะ key เป็น tuple
```

```python
a=float('-inf')
b=float('nan')
c=float('inf')
print([a,b,c]) # Python -> inf
print(json.dumps([a,b,c])) # JSON -> Infinity
```

```python
print(json.dumps({1:2,3:[4,[5,{6:7,8:9}]]},indent=4)) # indent คือ ความเยื้อง
```

```python
print(json.dumps({1:2,3:[4,[5,{6:7,8:9}]]}))
```

```python
dic={'d':1, 'I':2, 'c':3}
print(json.dumps(dic,sort_keys=True)) # ASCII ของ JSON ตัวพิมพ์ใหญ่มีค่าต่ำกว่าตัวพิมพ์เล็ก
```

```python
print(json.dumps({1:2,3:[4,[5,{6:7,8:9}]]})) #กั้นด้วย , และ : ที่ีตามด้วย 1 space ตามปกติ
print(json.dumps({1:2,3:[4,[5,{6:7,8:9}]]},separators=[',', ':']))   #กั้นด้วย , และ : ที่ไม่มี space
print(json.dumps({1:2,3:[4,[5,{6:7,8:9}]]},separators=[',  ', ':  '])) #กั้นด้วย , และ : ที่ตามด้วย 2 spaces
print(json.dumps({1:2,3:[4,[5,{6:7,8:9}]]},separators=['+ ', '| '])) #กั้นด้วย + และ | ที่ตามด้วย 1 space
```

```python
print(json.dumps(['ก', '๑']))
print(json.dumps(['ก', '๑'], ensure_ascii=False))
```

### คําสั่ง load
มีไฟล์ json แล้ว ต้องการอ้านขึ้นมาเป็น object ใน python

```python
with open('ab.json') as f:
	obj=json.load(f)
obj
```

### คําสั่ง loads
อ่านข้อความ json ที่เก็บไว้ในตัวแปรหรือเขียนข้อความ json เพื่อแปลง
เป็น object


```python
obj=json.loads('["a","b","e"]') #สังเกต loads มีเติม s ด้วย
print(obj)
```

## รูปแบบการแปลงของ JSON เป็น python
object -> dict

string -> str

array -> list

number(int) -> int

true -> True

number(float) -> float

false -> False

null -> None

```python
print(json.loads('["จ",15,3.2,null,false,true,{"4":15}]'))
```
