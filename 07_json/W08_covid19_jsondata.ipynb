API URL - ข้อมูลจะอัพเดตบ่อย

```python
import urllib.request
from pandas import json_normalize

url = 'https://covid19.ddc.moph.go.th/api/Cases/round-1to2-all'
fileobj = urllib.request.urlopen(url)
s = fileobj.read()
fileobj.close()
data = json.loads(s)
print(json_normalize(data))
```
