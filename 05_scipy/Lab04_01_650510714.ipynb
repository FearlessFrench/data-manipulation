```python
# อ่านข้อมูลจากไฟล์ data_0401.txt
def readFile(filename, mode='r'):
	fin=contents=None
	try:
		fin=open(filename,mode,encoding='utf-8')
		contents=fin.read()
	except:
		print("The file does not exist!")
	finally:
		if(fin is not None):
			fin.close()
	return contents

# เก็บลงในตัวแปรชื่อ data_list
data_list = readFile("data_0401.txt")
print(data_list)
```

```python
# data_list ต้องเป็นข้อมูลประเภท list of tuple
import csv
data_list = [('Name','Height(metre)','Weight(kg.)'),('นัทธี',1.71,62.5), ('Adam',1.75,84), ('Tommy',1.68,70.5), ('สาวิตรี',1.52,46),
             ('William',1.82,75.3), ('ภาวิกา',1.73,65.4), ('กมล',1.56,56.2), ('Kristy',1.67,72.3)]

# จากนั้นนาข้อมูล list ไปเขียนลงในไฟล์ data_650510714_0401.csv
csvfile = open('data_650510714_0401.csv','w',newline='')
obj = csv.writer(csvfile)

# โดยใช้ csv library และ writerow() method
for data_650510714_0401 in data_list:
	obj.writerow(data_650510714_0401)
csvfile.close()
```
