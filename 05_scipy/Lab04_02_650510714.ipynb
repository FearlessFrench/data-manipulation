```python
# ทำการอ่านไฟล์ data_650510714_0401.csv เพื่อรับค่าชื่อ(Name) ส่วนสูง (Height) และน้าหนัก (Weight)
def readFile(filename, mode='r'):
	fin=contents=None
	try:
		fin=open(filename,mode,encoding='utf-8')
		contents=fin.read()
	except:
		print("The file does not exist!")
	finally:
		if(fin is not None):
			fin.close()
	return contents

print(readFile("data_650510714_0401.csv"))
```
