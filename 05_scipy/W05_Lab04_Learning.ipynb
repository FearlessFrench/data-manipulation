```python
def writeFile(filename,contents,mode="wt"):
	fout=None
	try:
		fout=open(filename,mode,encoding='utf-8')
		fout.write(contents)
	finally:
		if(fout is not None):
			fout.close()
	return True
writeFile("fileout.txt","252,คิโมริ,plant,0.5,5.0")
```

```python
def writeFile(filename,contents,mode="wt"):
	with open(filename,mode) as fout:
		fout.write(contents)
writeFile("file2out.txt","252,คิโมริ,plant,0.5,5.0")
```

```python
# Version 2
def readFile(filename, mode='r'):
	fin=contents=None
	try:
		with open(filename,mode,encoding='utf-8') as fin:
			contents=fin.read()
	except:
			print("The file does not exist!")
	finally:
			if(fin is not None):
				fin.close()
	return contents
print(readFile("fileout.txt"))
```

```python
import urllib.request
def readWebPage(urlStr):
	assert(urlStr.startswith("http://"))  # ตรวจสอบว่า start with "http://" จริงมั้ย
	filehandler = contents = None
	try:
		filehandler=urllib.request.urlopen(urlStr)  # เปิดลิงค์
		contents=filehandler.read()
		contents=contents.decode('utf-8')
	finally:
		if(filehandler is not None):
			filehandler.close()
	return contents
print(readWebPage("http://www.pythons.org")[:200])
```

```python
# Shorter version
import urllib.request
import contextlib

def readWebPage(url):
	assert(url.startswith("http://"))
	with contextlib.closing(urllib.request.urlopen(url)) as fin:
		return fin.read()
print(readWebPage("http://www.pythons.org")[:200])
```

```python
import csv
persons=[('Ann',1,13),('Bill',4,30), ('Tome',2,15), ('Jim',4,23)]
csvfile=open('person.csv','w',newline='')
obj=csv.writer(csvfile)
for person in persons:
	obj.writerow(person)
csvfile.close()
```

```python
csvfile=open('person.csv', 'r',newline='')
obj=csv.reader(csvfile)
for line in obj:  # read แล้วก็ print ทีละบรรทัด
	print(line)
```
