จงเขียนโปรแกรมเพื่อทำการรับค่าเมตริกซ์ 2 ค่า ที่มีขนาดมิติคือ axb และ mxn จากนั้นทำการคำนวณค่าต่อไปนี้ (ก่อนทำการคำนวณให้ทำการตรวจสอบมิติว่าสามารถคำนวณได้หรือไม่ หากไม่สามารถคำนวณได้ ให้ทำการแสดงข้อความว่า “This calculation is wrong.”)

```python
import numpy as np
a = int(input("row = "))
b = int(input("column = "))
bf1 = np.random.randint(100,size=a*b)
print("list\n", bf1)
bf1 = bf1.reshape(a,b)
print("to matrix\n", bf1)
```

```python
m = int(input("row = "))
n = int(input("column = "))
bf2 = np.random.randint(100,size=m*n)
print("list\n", bf2)
bf2 = bf2.reshape(m,n)
print("to matrix\n", bf2)
```

```python
# ผลบวกของเมตริกซ์ทั้งสอง
try:
  sum = bf1 + bf2
  print(sum)
except ValueError:
  print("This calculation is wrong.")
```

```python
# ผลคูณของเมตริกซ์
try:
  mul = bf1 * bf2
  print(mul)
except ValueError:
  print("This calculation is wrong.")
```

```python
# รับค่าตัวแปร k โดย k เป็นเลขจำนวนเต็มใดๆ แล้วนำไปคูณกับเมตริกซ์ตัวแรกที่รับเข้ามา
k = int(input("multiplier = "))
try:
  res = k * bf1
  print(res)
except ValueError:
  print("This calculation is wrong.")
```
