จงเขียนโปรแกรมสำหรับรับค่าตัวแปร array_data ซึ่งเป็นข้อมูลชนิด ndarray ที่เป็นเมตริกซ์ขนาด nxn และทำการคำนวณค่าของตัวแปรดังต่อไปนี้ โดยสามารถเลือกใช้ method หรือฟังก์ชัน built-in

```python
import numpy as np
n = int(input("Enter dimension of matrix: "))

arr_data = np.random.randint(100,size=n*n)
print(arr_data)
print("---------------")
arr_data = arr_data.reshape(n,n)
print(arr_data)
print(type(arr_data))
```

```python
# บวกค่าข้อมูลแนวตั้งเก็บในตัวแปร sum_row
sum_row = arr_data.sum(axis=0)
print(sum_row)
sum_row = sum(sum_row)
print("Evaluate =", sum_row)
```

```python
# บวกค่าข้อมูลแนวนอนเก็บในตัวแปร sum_col
sum_col = arr_data.sum(axis=1)
print(sum_col)
sum_col = sum(sum_col)
print("Evaluate =", sum_col)
```

```python
# บวกค่าข้อมูลแนวทแยงมุมเก็บในตัวแปร sum_diagonal
# โดยค่าแรกคือ ผลบวกของค่าในแนวทแยงมุม จากมุมบนซ้ายไปยังล่างขวา
arr1 = np.diag(arr_data)
print(arr1)
# และค่าที่สองคือ ผลบวกของค่าในแนวทแยงมุม จากมุมบนขวาไปยังล่างซ้าย
arr2 = np.fliplr(arr_data)
arr2 = np.diag(arr2)
print(arr2)

sum_diagonal = sum(arr1) + sum(arr2)
print("Total =", sum_diagonal)
```

```python
# บวกค่าข้อมูลทุกตัวเก็บในตัวแปร sum_total
sum_total = sum(sum(arr_data))
print(sum_total)
```

```python
# ทำการหาเมตริกซ์ที่เกิดจากผลรวมของเมตริกซ์ที่กำหนดให้กับเมตริกซ์ Transpose
print(arr_data)
print("---------------")

arr3 = arr_data.transpose()
print(arr3)

print("---------------\nEvaluating...\n---------------")

try:
  result = arr_data+arr3
  print(result)
except ValueError:
  print("can not add matrix")
```
