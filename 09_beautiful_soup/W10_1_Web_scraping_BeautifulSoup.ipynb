```
url_naweb = 'https://hinaboshi.com/súpđẹp.html'
r = requests.get(url_naweb)
r.encoding = 'utf-8' # กำหนด encoding กันปัญหาการถอดรหัสผิดแบบ
sup = BeautifulSoup(r.text,'lxml')
```

ขอดูเนื้อหาภายในแท็ก

```
print(sup.h4) # มันจะเป็นแค่ find ไม่ใช่ find_all
# ได้ <h4 class="khokhwam">"ของอร่อยแบบนี้ไม่ว่าจะฤดูไหนมันก็อร่อยอยู่ดีแหละ"</h4>
print(sup.h4.contents)
# ได้ ['"ของอร่อยแบบนี้ไม่ว่าจะฤดูไหนมันก็อร่อยอยู่ดีแหละ"']
```

กรณีที่ต้องการสกัดเอาแค่ข้อความที่อยู่ในแท็กโดยไม่เอาส่วนตัวแท็กมาด้วย ใช้การต่อท้ายด้วย .text

```
print(sup.body.text)
```

นอกจากนี้ยังสามารถใช้ .children ซึ่งคล้าย .contents แต่จะออกมาเป็นในรูปแบบของอิเทอเรเตอร์ เหมาะสำหรับใช้ในวังวน for หรือถ้าจะแสดงทั้งหมดทันทีก็แปลงเป็นลิสต์

```
print(sup.find(id='wali1').children)
print(list(sup.find(id='wali1').children)) # อย่าลืมใส่ list ด้วยย
```

#### แท็กลูกและแท็กพ่อแม่
.contents หรือ .children จะแสดงแค่ส่วนประกอบที่อยู่ด้านในแท็กนั้นโดยตรง แต่ถ้าเป็น .descendants จะแสดงแท็กย่อยทั้งหมดไล่เรียงไปเลย โดยจะอยู่ในรูปของเจเนอเรเตอร์

```
print(len(list(sup.body.children)))
print(len(list(sup.body.descendants)))

a = sup.find(id='wali2').a # หาชิ้นส่วนที่มี id='wali2' แล้วค่อยไปดูที่ tag a
print(a)
for d in a.descendants:
    print(d.name)
```

แต่หากต้องการแสดงแท็กพ่อแม่ของแท็กนั้นๆ ก็ทำได้โดยเติม .parent

```
print(sup.body.parent.name)
print(sup.html.parent.name)
```

หรือถ้าใช้ .parents จะได้แท็กที่อยู่ขั้นสูงไล่ขึ้นไปเรื่อยๆทั้งหมด

```
print([p.name for p in sup.find(id='wali1').parents])
```

นอกจากนี้ยังมี .next_sibling, .previous_sibling, .next_siblings, .previous_siblings, .next_element, .previous_element, .next_elements, .previous_elements ที่เอาไว้ดูส่วนประกอบที่อยู่ถัดไปหรือก่อนหน้า

# **การใช้ regular expression**

## หาแท็กที่ชื่อชนิดมีตัว h ตามด้วยตัวเลข ได้แก่พวก h3 h4

```
import re

print(sup.find_all(re.compile(r'h\d'))) # ขึ้นต้นด้วย h ลงท้ายด้วยตัวเลข (\d)
```

## หาแท็กที่มี id ขึ้นต้นด้วย w

```
ww = sup.find_all(id=re.compile(r'w.+'))
```

```
re.findall('so+','so soooooo s see sooo sb soso ssso scbo')  #เพิ่มเติม ต้องมี o 1 ตัวเสมอ และสามารถซ้ำได้
```

```
re.findall('so*','so soooooo s see sooo sb soso ssso scbo')  #เพิ่มเติม มีหรือไม่มี o ก็ได้ และถ้ามี o ก็สามารถซ้ำได้ ตัวเสมอ และสามารถซ้ำได้
```

```
re.findall('s.+','so soooooo s see sooo sb soso ssso scbo')  #เพิ่มเติม
```

## การใช้งานที่มักพบบ่อยก็คือใช้เพื่อค้นหา url ของรูปภาพที่ต้องการจากนห้าเว็บแล้วทำการโหลด

```
url_naweb = 'https://hinaboshi.com/súpđẹp.html'
r = requests.get(url_naweb)
r.encoding = 'utf-8'
sup = BeautifulSoup(r.text,'lxml')
for klong in sup(class_='klonglek'):
    src = klong.img['src']
    chue_file = src.split('/')[-1]
    url_file = 'https://hinaboshi.com/'+src
    with open(chue_file,'wb') as f:
        f.write(requests.get(url_file).content)
```
