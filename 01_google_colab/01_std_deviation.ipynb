```python
# ให้เขียนฟังก์ชันเพื่อใช้สำหรับหาค่าความแปรปรวน (Standard Deviation)
# Sn = sqrt((1/N)*(sum((xi-u)**2)))
# N = จำนวน, u = ค่าเฉลี่ย, xi = ค่าแต่ละตัวใน list
def std_cal():
  a = [2.5, 3, 4.5, 6.7, 9, 10]
  b = [1, 1, 1, 1, 1, 1, 1, 1]
  c = [2.5, 2.5, 2.5, 3.7, 4, 4, 4, 0]
  N = len(a)
  u = sum(a)/N
  print(u)

  res = 0
  for x in a:
    res += (x - u)**2
  print(res)

  std = (res/N)**(1/2)
  print(std)

  import math
  total = math.ceil(std*100)/100
  print(total)

std_cal()
```
